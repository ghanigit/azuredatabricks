```python
#operators
#arithmetic operators
#assignmenet operators
#comparison operators
#logical operators
#identity operators
#membership operators
#bitwise operators

#arithmetic operators
print(10+20)
print(10-20)
print(10*20)
print(10/20)
print(9%2)   #remainder = 1 (4*2)
print(10**20)
print(9//2)  #floor division = 4 // Quotient = 4
print("end of arithmetic operators")

#assignment operators =,+=,-=,*=,/=,%=,//=,**=,&=,|=,^=,>>=,<<=,:=  
x=10
print(x)
x+=3  #x=x+3
print(x)    
x-=3  #x=x-3
print(x)
x*=3 #x=x*3
print(x)
x/=3 #x=x/3
print(x)
x%=3 #x=x%3
print(x)
x//=3 #x=x//3
print(x)
x**=3 #x=x**3
print(x)
print("end of assignment operators")
#comparison operators
# ==,!=,>,<,>=,<=
print(10==20)
print(10!=20)
print(10>20)
print(10<20)
print(10>=20)
print(10<=20)
print("end of comparison operators")
#logical operators
#and,or,not
print(10>20 and 10>30)
print(10>20 or 10>30)
print(not(10>20))
print("end of logical operators")
#identity operators
#is,is not
print(x is y)
print(x is not y)
print("end of identity operators")
#membership operators
#in,in not
fruits = ['apple', 'banana', 'cherry']
print("apple" in fruits)
print("apple" not in fruits)
print("end of membership operators")
```

A set is a collection of unique, unordered elements.

Key Characteristics:
✅ Unordered → No defined index or position
✅ No duplicates → Every element is unique
✅ Mutable → You can add or remove elements
🚫 Unindexed → You can’t access items by position

```python
fruits = {"apple", "banana", "cherry"}
print(fruits)

fruits.add("orange")
print(fruits)

#fruits.append("orange")
#print(fruits)

fruits.update(["orange", "mango", "grapes"])
print(fruits)

temp = list(fruits)
temp.append("pineapple")
fruits = set(temp)
print(fruits)


#fruits.remove("papaya") # if papaya is not present it will throw error
#print(fruits)

fruits.discard("papaya")            
print(fruits)   

fruits.discard("orange")            
print(fruits)

fruits.remove("pineapple")
print(fruits)

fruits.pop()
print(fruits)

fruits.clear()
print(fruits)

fruits.add("orange")
print(fruits)

duplicate_set= {1,2,3,1,2,3}
print(duplicate_set)

mixed_set = {1,0,2,2.5,"apple",False, True}
print(mixed_set)

boolean_set = {True, False, True, True}
print(boolean_set)
```

A dictionary in Python is a collection of key–value pairs.

It is:
Ordered (since Python 3.7+)
Mutable (can be changed)
Indexed by keys (not positions)
Does not allow duplicate keys

```python
student = {
    "name": "Alice",
    "age": 25,
    "course": "Data Science"
}

print(student)

print(student["name"])
print(student["age"])
print(student["course"])    

student["age"] = 26
print(student)

student["city"] = "New York"
print(student)

print(student.keys())
print(student.values())
print(student.items())
print(student.get("name1"))
print(student.get("city"))  

print(student.pop("name"))
print(student)

student.popitem()
print(student)

student.clear()     # it will clear the dictionary
print(student)
```

```python
#match  - similar switch case

day =input("enter day").lower()


match day:
  case "monday":
    print("monday")
  case "tuesday":
    print("tuesday")
  case "wednesday":
    print("wednesday")
  case "thursday":
    print("thursday")
  case "friday":
    print("friday")
  case "saturday":
    print("saturday")
  case "sunday":
    print("sunday")
  case _:
    print("invalid day")

```

```python
class person:
    def __init__(self,name,age):
        self.name = name
        self.age = age
    def greet(self):
        print(f"Hello, my name is {self.name} and I am {self.age} years old.")

p1=person("John",30)
p1.greet()
p2=person("Jane",25)
p2.greet()
```

```python
class Animal:
    def speak(self):
        print("Animal speaks")

class Dog(Animal):
    def bark(self):
        print("Dog barks")

class Cat(Animal):
    def meow(self):
        print("Cat meows")

d =Dog()
d.speak()
d.bark()

c =Cat()
c.speak()
c.meow()
```

```python
#polymorphism  - same method name different behaviour
class Bird:
    def sound(self):
        print("Bird chirps")

class Dog:
    def sound(self):
        print("Dog barks")

b =Bird()
b.sound()

d =Dog()
d.sound()

```
